```python
import librosa
import numpy as np
import librosa.display
```

```python
y, sr = librosa.load("sounds/heavy2.mp3")
```

```python
S = np.abs(librosa.stft(y))
```

```python
contrast = librosa.feature.spectral_contrast(S=S, sr=sr)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.subplot(2,1,1)
```

```python
librosa.display.specshow(librosa.amplitude_to_db(S,ref=np.max),y_axis='log')
```

```python
plt.subplot(2,1,2)
librosa.display.specshow(contrast, x_axis='time')
```

```python
y, sr = librosa.load("sounds/class_clip1.mp3")
S = np.abs(librosa.stft(y))
contrast = librosa.feature.spectral_contrast(S=S, sr=sr)
```

```python
plt.figure()
plt.subplot(2,1,1)
librosa.display.specshow(librosa.amplitude_to_db(S,ref=np.max),y_axis='log')
plt.subplot(2,1,2)
librosa.display.specshow(contrast, x_axis='time')
```

```python
y, sr = librosa.load("sounds/edm_clip1.mp3")
S = np.abs(librosa.stft(y))
contrast = librosa.feature.spectral_contrast(S=S, sr=sr)
```

```python
plt.figure()
plt.subplot(2,1,1)
librosa.display.specshow(librosa.amplitude_to_db(S,ref=np.max),y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Power spectrogram')

plt.subplot(2,1,2)
librosa.display.specshow(contrast, x_axis='time')
plt.colorbar()
plt.ylabel('Frequency bands')
plt.title('Spectral contrast')
plt.tight_layout()
```

```python
def plotSpectrum(filename):
    y, sr = librosa.load(filename)
    S = np.abs(librosa.stft(y))
    contrast = librosa.feature.spectral_contrast(S=S, sr=sr)
    plt.figure()
    plt.subplot(2,1,1)
    librosa.display.specshow(librosa.amplitude_to_db(S,ref=np.max),y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Power spectrogram')

    plt.subplot(2,1,2)
    librosa.display.specshow(contrast, x_axis='time')
    plt.colorbar()
    plt.ylabel('Frequency bands')
    plt.title('Spectral contrast')
    plt.tight_layout()
```

```python
plotSpectrum("sounds/class_clip1.mp3")
```
